```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import DataFrame,Series
import seaborn as sns
%matplotlib inline
```

```python
dframe=pd.read_excel("SampleData.xls")
```

```python
dframe.head()
```

```python
plt.figure(figsize=(18,14))
plt.subplot(211)
sns.countplot("Item",data=dframe)
plt.title("Item's Popularity")

plt.subplot(212)
sns.countplot("Item",data=dframe,hue="Region")
plt.title("Item's Popularity by Regions")
```

```python
dframe.groupby("Item").describe()
```

```python
sns.factorplot("Item","Total",data=dframe,palette="winter")
```
