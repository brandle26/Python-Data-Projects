## This is the analysis of the data that Meng-yi has provided. 

```python
import pandas as pd
from pandas import DataFrame,Series
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import cm
%matplotlib inline
```

```python
xls=pd.ExcelFile("SmartCityrRefined.xlsx")

MHP_df=xls.parse("Mean House Prices")
HAR_df=xls.parse("Housing Affordibility Ratio")
ER_df=xls.parse("Employment Rate")
AWE_df=xls.parse("Average weekly Earning")
```

```python
MHP_df.drop(["Belfast","Mansfield","Worthing","Bolton"],inplace="True")
MHP_df.drop([2011,2013],1).plot(kind="barh",figsize=(19,25))
plt.title("Mean House Prices in cities of UK",fontsize=18)
plt.ylabel("City",fontsize=15)
plt.xlabel("Mean House Prices in £1000s",fontsize=15)
```

### The graph above shows the mean house prices in cities of UK
It can be seen that, in average the house prices are highest in London and Oxford. 

```python
ER_df.drop("Belfast",inplace="True")
ER_df.drop([2011,2013],1).plot(kind="barh",figsize=(19,25))
plt.title("Employment Rate (%)",fontsize=18)
plt.ylabel("City",fontsize=15)
plt.xlabel("Employment Rate (%)",fontsize=15)
```

### The graph above shows the Employment Rate(%) in cities of UK
It can be seen that, highest employment rate is in warrington which is close to 80%. London also seems to have good employment rate of roughly 75%. Oxford also seesm to have  employment rate close to 75%.

```python
AWE_df.drop("Belfast",inplace="True")
AWE_df.plot(kind="barh",cmap="Spectral",figsize=(19,25))
plt.title("Average weekly Earning",fontsize=18)
plt.ylabel("City",fontsize=15)
plt.xlabel("Average weekly Earning",fontsize=15)
plt.show()   
```

### The graph above shows the Average weekly Earning in cities of UK
It can be seen that, highest weekly earning is for people in London. People in oxford also seem to be well off.

```python
HAR_df.drop("Belfast",inplace="True")
HAR_df.plot(kind="barh",figsize=(19,25))
plt.title("Housing Affordability Ratio",fontsize=18)
plt.ylabel("City",fontsize=15)
plt.xlabel("Housing Affordability Ratio",fontsize=15)
plt.show()
```

### The graph above shows housing Afforabnility ratio in cities of UK
Affordabiliy ratio compares earnings of eople with the housing cost of that city. In the graph higher affordability ration means more expensive city to live in. From the graph it can be seen that Oxford is the most expensive city to live in an London is the second most expensive city to live in.

## Conclusion

Oxford city is the most unaffordable place to live in. Alhthough the mean house price in oxford is quite high the average earning of the people in that city does not seems to be high enough to make it an affordable place. Close second seems to be London in terms of unafforable place to live.
